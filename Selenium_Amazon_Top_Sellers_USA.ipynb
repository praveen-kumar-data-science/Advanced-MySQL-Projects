```python
# Install Selenium
!pip install selenium

# Import necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

# Specify the URL and ChromeOptions
url = 'https://www.amazon.com/Best-Sellers-Clothing-Shoes-Jewelry-Mens-Boxer-Briefs/zgbs/fashion/1045718/ref=zg_bs_nav_fashion_4_1045710'

chrome_options = Options()
chrome_options.add_argument('--headless')  # Optional: Run Chrome in headless mode (no GUI)
chrome_options.add_argument('--no-sandbox')

# Start a Chrome browser session with system-installed Chrome
driver = webdriver.Chrome(options=chrome_options)

# Navigate to a website
driver.get(url)

# Add a delay to allow Chrome to start (you can adjust the delay as needed)
time.sleep(5)

# Interact with web elements (if needed)
# For example, to click a button by its ID:
# element = driver.find_element(By.ID, "button_id")
# element.click()
```

```python
# Get the body element of the webpage
body_element = driver.find_element(By.TAG_NAME, 'body')

# Extract the text content of the body element
page_text = body_element.text

# Print or use the extracted text
print(page_text)
```

```python
# Find all anchor elements (links) within the body
links = body_element.find_elements(By.TAG_NAME, 'a')

# Extract URLs from the links
urls = [link.get_attribute('href') for link in links]

urls
```

```python

# Find the target div element by its class name
target_div_class = 'a-cardui _cDEzb_card_1L-Yx'
target_div_class = 'p13n-desktop-grid'
target_div = driver.find_element(By.CLASS_NAME, target_div_class)

# Extract the text content of the target div
div_text = target_div.text

# Find the positions of '#1' and '#11'
pos1 = div_text.find('#1')
pos2 = div_text.find('#11')

extracted_text = div_text[pos1:pos2]
print("Extracted Text:", extracted_text.split('\n'))

# Find all anchor elements (links) within the target div
links = target_div.find_elements(By.TAG_NAME, 'a')

# Extract URLs from the links
urls = [link.get_attribute('href') for link in links]

# Print or use the extracted text and URLs
#print("Div Text:\n", div_text)
print("\nURLs:\n")
for i, url in enumerate(urls):
  if i%4 == 0 and i<=100:
    print(url)
```

```python
# Remove unwanted pattern using regex
import re
extracted_text = extracted_text.strip()
pattern = r"New Releases\nin .*\n" # Matches "New Releases in" followed by any characters until two newlines
extracted_text = re.sub(pattern, "", extracted_text)

print("Extracted Text (after removal):\n", extracted_text)
```

```python
all_texts = [word for word in extracted_text.split('\n') if word != '']
all_texts
```

```python
snos, item_names, review_cnts, prices = [],[],[],[]
url_links = []
for idx in range(len(all_texts)):
  if idx % 4 == 0:
    snos.append(all_texts[idx])
  elif idx % 4 == 1:
    item_names.append(all_texts[idx])
  elif idx % 4 == 2:
    review_cnts.append(all_texts[idx])
  else:
    prices.append(all_texts[idx])
```

```python
url_links = []
for i, url in enumerate(urls):
  if i%4 == 0 and i<=40:
    if 'gp/new-releases' in url:
      continue
    print(url)
    url_links.append(url)
```
